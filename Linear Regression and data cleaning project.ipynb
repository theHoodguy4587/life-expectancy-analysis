## Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Get the Data

```python
who=pd.read_csv('Life Expectancy Data.csv')
```

```python
who.head()
```

```python
who.describe()
```

```python
who.info()
```

```python
def clean_column_names(df):
    df.columns=df.columns.str.strip()
    df.columns=df.columns.str.replace(' ','_')
    df.columns=df.columns.str.lower()
    return df
who=clean_column_names(who)
```

## Check Missing Values

```python
who.isnull().sum()
```

```python
# Group by "Country" and check for completely null columns
all_null_columns=who.groupby('country').apply(lambda x:x.isnull().all())


fully_missing_columns={
    country:list(all_null_columns.columns[all_null_columns.loc[country]])
    for country in all_null_columns.index
    if all_null_columns.loc[country].any()
}
missing_columns_data=pd.DataFrame(list(fully_missing_columns.items()),columns=["country","Fully_missing_columns"])

# Display result
missing_columns_data
```

```python
single_record_countries=who.groupby('country').filter(lambda x: len(x) == 1)['country'].unique()
single_record_countries
```

```python
who.loc[who['country'].isin(single_record_countries)]
```

**Some countries have only one recorded entry, and their Life Expectancy value is missing. Since there is no way to accurately estimate or fill in the missing value for these cases, we have decided to remove these records from the dataset to maintain data reliability.**

```python
who = who.loc[~who['country'].isin(single_record_countries)]
```

## Fill missing values in numerical columns

```python
who['life_expectancy']=who.groupby('country')['life_expectancy'].transform(lambda x:x.fillna(x.median()))
who['adult_mortality']=who.groupby('country')['adult_mortality'].transform(lambda x:x.fillna(x.median()))
who['bmi']=who.groupby('country')['bmi'].transform(lambda x:x.fillna(x.median()))
```

```python
for col in ['polio','hepatitis_b','diphtheria']:
    who[col]=who.groupby('country')[col].transform(lambda x:x.fillna(x.mode()[0] if not x.mode().empty else x))
```

```python
for col in ['gdp','population','alcohol','total_expenditure']:
    who[col]=who.groupby('country')[col].transform(lambda x:x.interpolate(method='linear'))
```

```python
who.isnull().sum()
```

```python
who.dropna(inplace=True)
```

## Exploratory Data Analysis

### Top 10 countries with the highest life expectancy

```python
who.groupby('country')['life_expectancy'].mean().sort_values(ascending=False)[:10]
```

### Correlation Heatmap

```python
sns.heatmap(who.drop(['country','status','year'],axis=1).corr(),lw=0.5)
plt.show()
plt.title('Correlation Heatmap')
```

## Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
X=who.drop(['life_expectancy','country','status','year'],axis=1)
y=who['life_expectancy']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

## Training the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

### coefficients of the model

```python
print(lm.coef_)
```

## Predicting Test Data

```python
pred=lm.predict(X_test)
```

```python
sns.scatterplot(x=y_test,y=pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.show()
plt.xlabel('Actual Life Expectancy')
plt.ylabel('Predicted Life Expectancy')
```

## Residuals

```python
sns.histplot(y_test-pred,kde=True)
plt.show()
```

## Evaluating the Model

```python
from sklearn import metrics
```

```python
print('Mean Absolute Error (MAE): ',metrics.mean_absolute_error(y_test,pred))
print('Mean Squared Error (MSE): ',metrics.mean_squared_error(y_test,pred))
print('R-squared Score (R²): ',np.sqrt(metrics.root_mean_squared_error(y_test,pred)))
```
